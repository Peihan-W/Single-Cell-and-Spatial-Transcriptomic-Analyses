```r
library(Seurat)
library(harmony)
library(patchwork)
library(tidyverse)
library(magrittr)
library(future)

set.seed(1515)
options(future.globals.maxSize = 80000 * 1024^2)
plan("multicore", workers = 16)
plan()
```

```r
combined <- readRDS("/workdir/wangph/proj/20220418_TB/SC/5.seurat/v7-12sample-normal/1.3.merged-processed-20220812.rds")
```

```r
combined <- RunHarmony(combined, group.by.vars="sample", max.iter.harmony = 20) 
```

```r
combined@reductions$harmony
```

```r
ElbowPlot(combined, ndims = 50)
pc.num=1:30
combined <- RunTSNE(combined, reduction="harmony", dims=pc.num) %>% 
    RunUMAP(reduction="harmony", dims=pc.num) %>% 
    FindNeighbors(reduction="harmony", dims = 1:30) %>%
    FindClusters(resolution = c(0.5, 0.6, 0.8, 1, 1.2, 1.5, 1.8, 2, 2.5, 3))
```

```r
saveRDS(combined, file = "/workdir/wangph/proj/20220418_TB/SC/5.seurat/v7-12sample-normal/1.4.merged-harmony-processed-20220811.rds")
```
